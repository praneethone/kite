```python
# Write a function that returns the lesser of two given numbers if both numbers are even, but returns the greater if one or both numbers are odd.
def lesser_of_two(a,b):
    if a%2==0 and b%2==0:
        return min(a,b)
    else:
        return max(a,b) 
lesser_of_two(4,6)    
```

```python
def count_upper_lower(s: str):
    upper = sum(1 for i in s if i.isupper())
    lower = sum(1 for i in s if i.islower())
    return upper, lower
print(count_upper_lower("AbC uudeF!")) 
```

```python
def count_upper_lower(s):
    upper_count = 0
    lower_count = 0

    for char in s:
        if char.isupper():         
            upper_count += 1
        elif char.islower():        
            lower_count += 1

    return upper_count, lower_count
```
